```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name = "mi_collection")
```

```python
collection.add(
    documents = [
        "This document is about İzmir",
        "This document is about Finlandiya"
    ],
    ids = ["id1", "id2"]
)
```

```python
all_docs = collection.get()
all_docs
```

```python
documents = collection.get(ids=["id1"])
documents
```

```python
results = collection.query(
    query_texts =  ["Query is about İzmir meatball"],
    n_results = 2
)
print(results)
```

```python
# Dokümanları silme

collection.delete(ids = all_docs["ids"])
collection.get()
```
